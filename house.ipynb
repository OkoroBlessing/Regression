```python
import tensorflow as tf
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, InputLayer
```

```python
data=pd.read_csv('USA_Housing.csv')
data.head()
```

```python
data.info()
```

```python
data=data.drop('Address', axis=1)
data.head()
```

```python
x =data.drop('Price', axis=1)
y = data['Price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=0.2,random_state=2024) #use stratify in unbalanceddata 
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
x_train_scaled
```

```python
model = Sequential()
model.add(InputLayer(input_shape=(5,)))
# model.add(Dense(units=32, activation="relu",input_shape=(5,)))
model.add(Dense(units=16, activation="relu"))
model.add(Dense(units=64, activation="relu"))
model.add(Dense(units=128, activation="relu"))
model.add(Dense(units=1, activation ="linear"))
```

```python
model.summary()
```

```python
model.compile( loss ='mean_squared_error',metrics=['mae'])
```

```python
model.fit(x=x_train_scaled, y=y_train,epochs=50, verbose=1,validation_split=0.1)
```

```python
from sklearn.metrics import r2_score
```

```python
prediction = model.predict(x_test_scaled)
prediction
```

```python
r2_score(y_test,prediction)
```
